Add column by assigning series to dataframe

```python
car_sales["Seats"] = pd.Series([5,5,5,5,5])
car_sales["Seats"] = car_sales.Seats.fillna(5)
car_sales["Seats"] = car_sales.Seats.astype(int)
car_sales
```

Column from python list

```python
fuel_economy = [7.5, 9.2, 5., 9.6, 8.7, 4.7, 7.6, 8.7, 3.0, 4.5]
car_sales["Fuel per 100Km"] = fuel_economy
car_sales
```

```python
car_sales["Total fuel used (L)"] = car_sales["Odometer (KM)"] / 100 * car_sales["Fuel per 100Km"]
car_sales
```

```python
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales[["Make","Price2"]]
```

```python
car_sales_shuffeled = car_sales.sample(frac=1)
```

```python
car_sales_shuffeled = car_sales_shuffeled.reset_index(drop=True)
car_sales_shuffeled
```

```python
car_sales_shuffeled["Odometer (Mile)"] = car_sales_shuffeled["Odometer (KM)"].apply(lambda x: x / 1.6)
car_sales_shuffeled
```
